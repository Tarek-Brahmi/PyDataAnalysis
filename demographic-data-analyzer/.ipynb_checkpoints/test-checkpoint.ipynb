```python
# What is the average age of men?
df.loc[df['sex']=="Male"].mean()["age"]
```

```python
# What is the percentage of people who have a Bachelor's degree?
x=df.loc[df['education']=="Bachelors","education"]
(len(x)/len(df))*100
```

```python
# `Bachelors`, `Masters`, or `Doctorate`
higher_education=df.loc[(df['education']=="Bachelors") |(df['education']=="Masters") |(df['education']=="Doctorate" )]
```

```python
len(higher_education)/len(df)*100
```

```python
lower_education=df[(df['education']!="Bachelors") & (df['education']!="Masters") & (df['education']!="Doctorate") ]
```

```python
lower_education
```

```python
# percentage with salary >50K
higher_education_rich=df.loc[(df["salary"]==">50K")].loc[(df['education']=="Bachelors") |(df['education']=="Masters") |(df['education']=="Doctorate" )]
```

```python
len(higher_education_rich)/len(higher_education)*100
```

```python
lower_education_rich=df.loc[(df["salary"]==">50K")].loc[(df['education']!="Bachelors") & (df['education']!="Masters") & (df['education']!="Doctorate") ]
```

```python
lower_education_rich
```

```python
(len(df.loc[(df["salary"]==">50K")].loc[(df['education']!="Bachelors") & (df['education']!="Masters") & (df['education']!="Doctorate") ])/len(lower_education))*100
```

```python
# What is the minimum number of hours a person works per week (hours-per-week feature)?
min_work_hours=df["hours-per-week"].min()
```

```python
min_work_hours
```

```python
# What percentage of the people who work the minimum number of hours per week have a salary of >50K?
num_min_workers=df.loc[(df["salary"]==">50K")].loc[(df["hours-per-week"]==df["hours-per-week"].min())]
```

```python
num_min_workers
```

```python
rich_percentage=len(num_min_workers)/len(df.loc[(df["hours-per-week"]==df["hours-per-week"].min())])*100
```

```python
rich_percentage
```
